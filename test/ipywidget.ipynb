```python
%matplotlib ipympl
from ipywidgets import interact
import matplotlib.pyplot as plt
import tifffile

img = tifffile.imread('../data/181228_CDX2_9s_c48_n073.tif')
fig, ax = plt.subplots()
aximg = plt.imshow(img[0])
plt.axis('off')

@interact(t=(0, img.shape[0]-1, 1),cmap=['viridis','jet','gray'])
def display(t,cmap):
    aximg.set(data=img[t], cmap=cmap)
    ax.set_title(f'frame {t}')
    return 
```
